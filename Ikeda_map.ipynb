# **Ikeda**

ikeda_map.PNG

```
from matplotlib import pyplot as plt
import numpy as np

# Parameters
u =  0.9

# ikeda Math operation 
def ikeda(x, y): 
    tn = 0.4 - 6/(1+x**2 +y**2)
    xn = 1 + u*(x*np.cos(tn) - y*np.sin(tn))
    yn = u*(x*np.sin(tn) + y*np.cos(tn))
    return (xn, yn)

# Looping over ikeda function, to creat X and Y
X, Y = [], []
x, y = 0, 0
L = 100000
for i in range(L):
    x, y = ikeda(x, y)
    X, Y = X + [x], Y + [y]

# X vs Y Scatter Plot
fig, ax = plt.subplots(figsize=(10,10))
ax.scatter(X, Y, color='blue', s=0.1)
plt.xlabel('$Re(E_n)$', fontsize=15)
plt.ylabel('$Im(E_n)$', fontsize=15)
plt.tick_params(labelsize=15)
plt.savefig('ikeda_map.png')
```
